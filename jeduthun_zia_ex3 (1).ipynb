```python
###PROBLEM 1

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns 

%matplotlib inline

# prepare and read in data
url = 'https://raw.githubusercontent.com/jbebop/csv/main/WHR20_DataForTable2.1.csv'
df = pd.read_csv(url, header = 0,
                usecols = ['life_ladder'])  

dfs = df.sort_values('life_ladder', ascending = False) 
print(dfs.describe()) #for reference 

fig, ax = plt.subplots(figsize=(10, 6)) 
sns.kdeplot(dfs.life_ladder, ax=ax, 
                #bw=0.5,
                #bw=1,
                #bw=2,
                bw=0.2, 
                shade=True,            
                alpha=0.6,   #Gaussian kernel          
                cut=0                 
               )                        

sns.rugplot(dfs.life_ladder, ax=ax, height=0.04)    # shows location of individual observations at the bottom.

##matplotlib
#plt labels
ax.set_xlabel('Life Satisfaction Rating')
ax.set_ylabel('Density')
ax.set_title('Life Satisfaction Ratings from the World Health Report')
#plt visualizations
ax.grid(axis='y')           
ax.set_axisbelow(True)      
plt.box(False)          
```
